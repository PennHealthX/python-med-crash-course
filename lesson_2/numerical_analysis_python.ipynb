## Using External Code: Import Statements

Python comes with many built-in tools (called **modules** or **libraries**) that add extra functionality. Instead of writing everything from scratch, you can **import** these tools and use them!

### Why Use Imports?

Think of modules like apps on your phone. Your phone has basic features, but you download apps to do specific things (social media, games, navigation). Similarly, Python has basic features, but you import modules to do specialized tasks like:

- Math calculations (beyond basic +, -, *, /)
- Working with dates and times
- Reading/writing files
- Analyzing data
- Making graphs
- And much more!

**NumPy**, **Pandas**, and **Matplotlib** are external libraries that we will need to import.

### Basic Import

```python
# Import the math module
import math

# Now we can use functions from the math module
print(math.sqrt(16))        # Square root of 16 = 4.0
print(math.pi)              # The value of pi
print(math.pow(2, 3))       # 2 to the power of 3 = 8.0
print(math.floor(4.7))      # Round down = 4
print(math.ceil(4.3))       # Round up = 5
```

Notice the pattern: `modulename.function()`. After importing, you use the module name followed by a dot, then the function name.

### Importing Specific Functions

If you only need specific functions, you can import just those:

```python
# Import only sqrt and pi from math
from math import sqrt, pi

# Now you can use them without "math."
print(sqrt(25))    # 5.0
print(pi)          # 3.141592...
```

### Using Aliases (Nicknames)

Sometimes module names are long. You can give them shorter nicknames:

```python
# Import with a nickname (very common in data science!)
import math as m

print(m.sqrt(64))  # 8.0
```

### Common Built-in Modules

Here are some useful modules that come with Python (no installation needed):

#### 1. random - Generate random numbers

```python
import random

# Random integer between 1 and 10
print(random.randint(1, 10))

# Random decimal between 0 and 1
print(random.random())

# Choose a random item from a list
colors = ["red", "blue", "green", "yellow"]
print(random.choice(colors))
```

#### 2. datetime - Work with dates and times

```python
from datetime import datetime

# Get current date and time
now = datetime.now()
print("Current date and time:", now)

# Get just the date
print("Today's date:", now.date())

# Get just the time
print("Current time:", now.time())
```

#### 3. statistics - Basic statistical calculations

```python
import statistics

data = [10, 20, 30, 40, 50]

print("Mean:", statistics.mean(data))      # Average
print("Median:", statistics.median(data))  # Middle value
print("Stdev:", statistics.stdev(data))    # Standard deviation
```

## Numpy

NumPy is a python library used for efficient computations. We often use NumPy to manipulate data stored within larger data tables. NumPy's efficiency relies on storing data that is all of the same type. Here, we will explore NumPy functions by applying them to 1D arrays. In the Pandas section, we will explore more advanced and practical uses of NumPy. 

```python
import numpy as np

# Create arrays with medical information
age = np.array([32, 15, 23, 66])
heart_rate = np.array([66, 120, 176, 77])
systolic_bp = np.array([140, 110, 80, 97])

# Use numpy functions to create statistics
average_age = np.mean(age)
average_hr = np.mean(heart_rate)
average_sbp = np.mean(systolic_bp)

print(f"The average age, heart rate, and systolic BP are {average_age}, {average_hr}, and {average_sbp}.")
```

```python
# Similarly to a list, we can index into numpy arrays
first_2_ages = age[:2]
last_2_ages = age[-2:]
print(first_2_ages)
print(last_2_ages)
```

We often save the outcome list as a separate array. This allows us to look at all the variables affecting the outcome separately. Blinding ourselves to the outcome will be helpful when we develop prediction models.

```python
mortality = np.array([0, 0, 1, 1])
dead = np.where(mortality == 1)[0]

for i in dead:
    print(f"Patient {i} is deceased.")
```

We can also use NumPy to change numerical variables into categorical variables. Suppose we want to know which patients are hypertensive. We can use NumPy to set a condition (systolic blood pressure >= 130) and apply that rule to each patient/element of the array, marking that patient as 'Positive' or 'Negative' for hypertension.

```python
hypertension = np.where(systolic_bp >= 130, 'Positive', 'Negative')
print(systolic_bp)
print(hypertension)
```

### Exercise 1: For patients with heart disease, determine the fraction of people who are current smokers.

```python
# Select for patients with heart disease
heart_disease = stroke_data.loc[stroke_data["heart_disease"]==1, ["smoking_status"]]

# Find the number of current smokers and patients with heart disease
num_heart_disease = len(heart_disease)
curr_smokers = len(heart_disease[heart_disease["smoking_status"]=="smokes"])

# Calculate the fraction of people with heart disease who smoke
fraction = curr_smokers / num_heart_disease
print(f"There are {num_heart_disease} patients with heart disease, of which {curr_smokers} smoke.")
print(f"The fraction of patients with heart disease who smoke is {fraction}.")
```

### Exercise 2: Calculate the average glucose level for people living in urban vs. rural areas. 

```python
# Find patients who live in urban and rural areas
stroke_data["Residence_type"].value_counts()
urban = stroke_data.loc[stroke_data["Residence_type"]=="Urban", :]
rural = stroke_data.loc[stroke_data["Residence_type"]=="Rural", :]

# Calculate the average glucose level for each group
print(f"The average glucose level for patients living in an urban setting is {urban['avg_glucose_level'].mean():.2f}")
print(f"The average glucose level for patients living in a rural setting is {rural['avg_glucose_level'].mean():.2f}")
```

### Alternative: Group By

## Matplotlib

Matplotlib is a library that is commonly used to create graphs. This can be used to visually present the findings that you derived from using Pandas functions.

```python
import matplotlib.pyplot as plt
```

```python
stroke_data.head()
```

Categorical data with matplotlib.

```python
import matplotlib.pyplot as plt

# Aggregate: count stroke cases by smoking status
stroke_counts = stroke_data.groupby("smoking_status")["stroke"].sum()  # number of stroke patients per category

# Plot
plt.figure(figsize=(6,4))
stroke_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Stroke Patients by Smoking Status")
plt.ylabel("Number of Stroke Patients")
plt.xlabel("Smoking Status")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

Numerical data with matplotlib.

```python
plt.figure(figsize=(8,5))

# Filter data for each group
stroke = stroke_data[stroke_data['stroke'] == 1]['avg_glucose_level']
no_stroke = stroke_data[stroke_data['stroke'] == 0]['avg_glucose_level']

# Plot histograms with transparency
plt.hist(no_stroke, bins=30, alpha=0.5, label='No Stroke', color='skyblue', edgecolor='black')
plt.hist(stroke, bins=30, alpha=0.7, label='Stroke', color='salmon', edgecolor='black')

plt.title("Glucose Distribution: Stroke vs. Non-Stroke")
plt.xlabel("Average Glucose Level (mg/dL)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

# Filter data for each group
stroke = stroke_data[stroke_data['stroke'] == 1]['age']
no_stroke = stroke_data[stroke_data['stroke'] == 0]['age']

# Plot histograms with transparency
plt.hist(no_stroke, bins=30, alpha=0.5, label='No Stroke', color='skyblue', edgecolor='black')
plt.hist(stroke, bins=30, alpha=0.7, label='Stroke', color='salmon', edgecolor='black')

plt.title("Age Distribution: Stroke vs. Non-Stroke")
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
